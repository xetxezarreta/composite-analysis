```python
import pandas as pd
import os
from glob import glob
from tsfresh.feature_extraction import feature_calculators

files = glob('D:\master\LIP\composite-analysis\data\*.csv')

totals = ['Total_PorosityQuantity', 'Total_PorosityQuality', 'Total_UnfilledZones', 'Total_FillingQuality', 'TOTAL_QUALITY']

output_file = 'extracted_data.csv'

features = {
    feature_calculators.abs_energy, # The absolute energy of the time series which is the sum over the squared values.
    feature_calculators.count_above_mean, # The number of values in x that are higher than the mean of x.
    feature_calculators.count_below_mean, # The number of values in x that are lower than the mean of x.
    feature_calculators.kurtosis, # The kurtosis of x (calculated with the adjusted Fisher-Pearson standardized moment coefficient G2).
    feature_calculators.maximum, # The highest value of the time series x.
    feature_calculators.minimum, # The lowest value of the time series x.
    feature_calculators.mean, # The mean of x.
    feature_calculators.skewness, # The sample skewness of x (calculated with the adjusted Fisher-Pearson standardized moment coefficient G1).
    feature_calculators.standard_deviation, # The standard deviation of x.
    feature_calculators.variance # The variance of x        
}

def get_df_features(df, series):
    df_features = pd.DataFrame()
    for column_name in series:
        for feature in features:
            column = df[column_name]
            name = column_name + '_' + feature.__name__
            value = feature(column)
            df_features[name] = pd.Series(value)
    return df_features      


output_file = 'extracted_data.csv'

for i, file in enumerate(files):
    if not file.endswith('.csv'):
        continue
    print("Extracting file: " + str(i) + " (" + file + ")")
    df = pd.read_csv(file)
    target = df.TOTAL_QUALITY.unique()[0]
    df = df.drop(axis=1, columns=totals)
    series = [col for col in df if col.endswith(('Time', 'Flow rate', 'Pressure'))]  
    features_df = get_df_features(df, series)
    features_df['Target'] = target
    
    if not os.path.isfile(output_file):
        features_df.to_csv(output_file)
    else:
        features_df.to_csv(output_file, mode='a', header=False)
```
